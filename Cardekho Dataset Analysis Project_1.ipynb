## Step 6 : Data Visualization

```python
# Assigning data to df variable
df = pd.DataFrame(data)

# Creating a pie chart for car owners
df['owner'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('Car Owners',weight='bold',fontsize=14)
plt.legend(loc='upper right', fontsize=12, shadow=True,bbox_to_anchor=(2,1))  # Set legend properties
plt.show()
```

```python
# Creating pie chart for dealers
colors = ['gold', 'yellowgreen', 'lightcoral']
df['seller_type'].value_counts().plot(kind='pie',autopct='%1.1f%%',colors=colors,startangle=90)
plt.title('Dealers',weight='bold',fontsize=14)
plt.legend(loc='upper right', fontsize=12, shadow=True,bbox_to_anchor=(2,1))
plt.xticks(rotation=140)
plt.show()
```

```python
# Creating bar chart for top 10 cars sold
plt.figure(figsize=(6,4))
df['name'].value_counts().head(10).plot(kind='bar',color='#FF0000',fontsize=10,edgecolor='black',width=0.8)
plt.xlabel('Cars',weight='bold',fontsize=14)
plt.ylabel('Count',weight='bold',fontsize=14)
plt.title('Top 10 Cars sold',weight='bold',fontsize=14)
plt.xticks(rotation=90)
plt.show()
```

```python
# Creating  bar chart of top 10 car brand sold
plt.figure(figsize=(6,4))
df['brand'].value_counts().head(10).plot(kind='bar',color='#1f77b4',edgecolor='black',width=0.8)
plt.xlabel('Cars Brand')
plt.ylabel('Count')
plt.title('Top 10 Cars brand sold')
plt.xticks(rotation=90)
plt.show()
```

```python
# Creating histogram of selling price distribution
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
plt.hist(df['selling_price'],color='#ff7f0e',edgecolor='black')
plt.title('Selling Price Distribution')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')

# Creating histogram of kilometer driven distribution
plt.subplot(2, 2, 2)
plt.hist(df['km_driven'], color='#ff7f0e',edgecolor='black')
plt.title('Kilometers Driven Distribution')
plt.xlabel('Kilometers Driven')
plt.ylabel('Frequency')

# Creating histogram of mileage distribution
plt.subplot(2, 2, 3)
plt.hist(df['mileage(km/ltr/kg)'],color='#ff7f0e', edgecolor='black')
plt.title('Mileage Distribution')
plt.xlabel('Mileage (km/ltr/kg)')
plt.ylabel('Frequency')

# Creating histogram of engine distribution
plt.subplot(2, 2, 4)
plt.hist(df['engine'], color='#ff7f0e',edgecolor='black')
plt.title('Engine Distribution')
plt.xlabel('Engine')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Creating bar chart of cars sold in a particular year
plt.figure(figsize=(8,4))
df['year'].value_counts().sort_index().plot(kind='bar',color='#2ca02c',edgecolor='black',width=0.8)
plt.title('Count of Cars by Year',weight='bold',fontsize=14)
plt.xlabel('Year',weight='bold',fontsize=14)
plt.ylabel('Count',weight='bold',fontsize=14)
```

```python
# Creating bar chart of number of cars sold by fuel type
plt.figure(figsize=(6,4))
df['fuel'].value_counts().plot(kind='bar',color='#f34089',edgecolor='black',width=0.8)
plt.title('Count of Cars by Fuel Type',weight='bold',fontsize=14)
plt.xlabel('Fuel Type',weight='bold',fontsize=14)
plt.ylabel('Count',weight='bold',fontsize=14)
plt.show()
```

```python
# Creating bar chart of number of cars sold by transmission type
plt.figure(figsize=(6,4))
df['transmission'].value_counts().plot(kind='barh',color='#2ca02c',edgecolor='black')
plt.title('Count of Cars by transmission type',weight='bold',fontsize=14)
plt.xlabel('Count of cars',weight='bold',fontsize=14)
plt.ylabel('Transmission type',weight='bold',fontsize=14)
plt.show()
```

```python
# Creating box chart of mileage vs fuel type
# Create a dictionary to store mileages for each fuel type
mileage_by_fuel_type = {}
for fuel_type in data['fuel'].unique():
    mileage_by_fuel_type[fuel_type] = data[data['fuel'] == fuel_type]['mileage(km/ltr/kg)']

# Create a list of mileages grouped by fuel type
mileage_data = [mileage_by_fuel_type[fuel_type] for fuel_type in sorted(mileage_by_fuel_type.keys())]

# Create the box plot
plt.figure(figsize=(12, 6))
plt.boxplot(mileage_data, labels=sorted(mileage_by_fuel_type.keys()))
plt.title('Fuel Type vs Mileage')
plt.xlabel('Fuel Type')
plt.ylabel('Mileage (km/ltr/kg)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Creating line chart to know trend of selling price over years
plt.figure(figsize=(8,4))
df.groupby('year')['selling_price'].mean().plot(marker='o')
plt.xlabel('Year',weight='bold',fontsize=14)
plt.ylabel('Average Selling Price',weight='bold',fontsize=14)
plt.title('Trend of Selling Price over Years',weight='bold',fontsize=14)
plt.show()
```
